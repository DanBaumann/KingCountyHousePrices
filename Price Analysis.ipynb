# Price analysis 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import geopandas
import folium
import matplotlib.pyplot as plt
from scipy.stats import skew
from scipy.stats.stats import pearsonr
%matplotlib inline
```

```python
df = pd.read_csv('kc_house_data.csv')
```

Let's see a visual represntation of house prices in the data set using a histogram plot

```python
# histogram
f, ax = plt.subplots(figsize = (14, 6))
fig = sns.distplot(df['price'], bins = 50, kde = False)
# there is clearly some positive skewness which suggest deviation from normal distribution
```

```python
display("Skewness: %f" % df['price'].skew())
display("Kurtosis: %f" % df['price'].kurt())
# a kurtosis of 3 indicates normal distribution, so house prices has a far greater distribution around median than a normal distribution
# it also means that mean != median
```

```python
# correlation matrix
correlation_matrix = df.corr()
f, ax = plt. subplots(figsize = (15,12))
sns.heatmap(correlation_matrix, vmin = -1, vmax = 1, cmap = 'coolwarm', annot = True, square = True)
```
