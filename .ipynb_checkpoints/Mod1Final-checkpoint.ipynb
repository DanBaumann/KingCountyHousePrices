# Exploring Null Values for Waterfront

The dataset includes 2376 null values for waterfront. We wanted to explore where these houses were located on a map. 
To do this we imported a shapefile using geopandas

```python
king_county_map = gpd.read_file('/Users/danbaumann/Downloads/tl_2017_53033_roads/tl_2017_53033_roads.shp')
# We retrieved a shapefile from google for the King County area 
```

```python
wf_coordinates = df.loc[:, ['waterfront', 'lat', 'long']]
# We created a separate dataframe with only waterfront, lat, and long as variables
na_dropped_coordinates = wf_coordinates.loc[(wf_coordinates.waterfront.isna()) & (df.lat) & (df.long)]
# then dropped all observations which had non-null observations
na_dropped_coordinates.head()
```

```python
fig, ax = plt.subplots(figsize = (15,8))
king_county_map.plot(ax = ax)
crs = {'init': 'epsg:4326'}
# We used a standard coordinate reference system and zipped the longitude and 
#     latitude of observations where waterfront was given a null value
# This gives us t
```

```python
geometry = [Point(xy) for xy in zip(na_dropped_coordinates['long'], na_dropped_coordinates['lat'])]
geometry[:3]
geo_df = gpd.GeoDataFrame(na_dropped_coordinates, crs = crs, geometry = geometry)
geo_df.head()
# We created a GeoDataFrame which gave us a Point using the above CRS
```

```python
fig, ax  = plt.subplots(figsize = (15,15))
king_county_map.plot(ax = ax, alpha = 0.4, color = 'grey')
geo_df.plot(ax = ax, markersize = 10, color = 'red', marker = 'o', label = 'Location of Null Waterfront')
plt.legend(prop = {'size': 15})
```

We found that there were a large amount of observations that were given null values for waterfront, however as depicted by the map above, a considerable amount of properties are situated around the waterfront. This will be useful for our data cleaning later on in the project.

## Data Cleaning

It appears that most categories are not missing any observations.


Date is an object, this is ok
sqft_basement should be converted into an integer like other sqft measurements

```python
df.sqft_basement.value_counts().head()
```

# there are 454 houses with '?' as basement square footage
# possibility of dropping these values, or changing to 0
# worth checking if sqft_living is different to sqft_above, which would indicate there is a basement


# df.loc[(df.sqft_basement == '?')]?
# df.applymap(lambda x_id: df.iloc[x_id, :].sqft_living - x.iloc[x_id, :].sqft_above if x == '?' else x.sqft_basement)

```python
def compute_sqft_basement(observation):
    if observation.sqft_basement == '?':
        observation.sqft_basement = observation.sqft_living - observation.sqft_above
    return observation
```

```python
df = df.apply(compute_sqft_basement, axis=1)
# df.head()
df.sqft_basement.value_counts().head()
# we got rid of the '?' now, but need to convert it from object to float
```

```python
df.sqft_basement = pd.to_numeric(df.sqft_basement)
df.sqft_basement.value_counts().head()
df.sqft_basement = df.sqft_basement.astype(np.int64)
df.info()
df.isna().sum()
```
