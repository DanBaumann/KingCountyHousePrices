# Question 2. *What is the effect that living on the waterfront has on property price?*

```python
wf_avg = df.groupby('waterfront').mean()
# The above variable assignment groups the dataframe by waterfront
```

Waterfront presence has been used as a binary variable. A value of 1 indicates that the property is situated on the waterfront, with the alternative 0, meaning that the property is not on the waterfront. There is a significant difference in mean price between waterfront and non-waterfront properties. Properties on the waterfront have an average value nearly 3 times greater. While the number of bedrooms is similar, square footage of living space and lot space differ greatly. However, the disparity in price cannot be due to these variables alone.

This suggests that waterfront may have a large impact on price. In order to further our investigation, we thought that isolating zipcodes would be an appropriate step in comparing prices of waterfront versus non-waterfront properties. By adding this restriction of location, we have taken a step in reducing bias in our findings. Our assumption is that properties that are located within a certain zipcode are more likely to be similar in size and prices. By doing this, we are better able to extrapolate any findings and say them with more certainty.

## (4) Data Cleaning

It appears that most categories are not missing any observations.


Date is an object, this is ok
sqft_basement should be converted into an integer like other sqft measurements

```python
df.sqft_basement.value_counts().head()
# there are 454 houses with '?' as basement square footage
# possibility of dropping these values, or changing to 0
# worth checking if sqft_living is different to sqft_above, which would indicate there is a basement
```

```python
# df.loc[(df.sqft_basement == '?')]?
# df.applymap(lambda x_id: df.iloc[x_id, :].sqft_living - x.iloc[x_id, :].sqft_above if x == '?' else x.sqft_basement)
# some code we struggled with before finally working out how to convert sqft_basement observations with '?'
```

```python
def compute_sqft_basement(observation):
    if observation.sqft_basement == '?':
        observation.sqft_basement = observation.sqft_living - observation.sqft_above
    return observation
# the above function changes observations with '?' as a value to their correct value. sqft_living - sqft_above should
#     equal sqft_basement
```

```python
df = df.apply(compute_sqft_basement, axis=1)
# this is the computation of our function with our missing values 
df.sqft_basement.value_counts().head()
# we got rid of the '?' now, but need to convert it from object to float
```

```python
df.sqft_basement = pd.to_numeric(df.sqft_basement)
df.sqft_basement.value_counts().head()
df.sqft_basement = df.sqft_basement.astype(np.int64)
df.info()
df.isna().sum()
```

# (6) Predictive Modelling 

The underlying objective of this project has been to verify any relationship the given variables have on the prices of houses in King County. Now that data has been cleaned and we have shown some data analysis, we can begin our predictive modelling of house prices. 

Owing to our EDA, we saw a distinct trend in how property features changed over time. In the last 25 years, square footage of living space increased, but total lot size decreased. For this reason, we decided to create separate regressions based on the age of the house. With these separate models, it will be easier to recommend how to invest or perceive property features for houses that were built in different time periods. 

# 6.1 *A regression model for houses built within the last 25 years*

```python
df_last_25_years = df.loc[df['house_age'] < 25]
df_last_25_years.shape
# new DataFrame contains 5426 observations 
```

```python
correlation_matrix = df_last_25_years[['price', 'sqft_living', 'waterfront', 'view', 'lat', 'long']].corr()
f, ax = plt. subplots(figsize = (15,12))
sns.heatmap(correlation_matrix, cmap = 'RdYlGn_r', annot = True, square = True)
# Correlation matrix with new DataFrame helps us to avoid colinearity 
# comment more on this and why we are going to choose certain variables 
plt.savefig('correlation_matrix.png', dpi=300, bbox_inches='tight')
```

Want to use sqft_living, waterfront, view, lat, long. We picked these variables because they are all somewhat strongly correlated with our target variable price. On top of this, we checked for colinearity and the highest correlation between variables was waterfront and view, which had a PearsonR correlation of 0.37. Hopefully this selection will avoid problems of multicolinearity. 

```python
outcome = 'price'
x_cols =  ['sqft_living', 'waterfront', 'view', 'lat', 'long']
predictors = '+'.join(x_cols)
formula = outcome + '~' + predictors 
model1 = ols(formula = formula, data = df_last_25_years).fit()
model1.summary()
```

```python
residuals = model1.resid
fig = sm.graphics.qqplot(residuals, dist=stats.norm, line='45', fit=True)
# Our Q-Q plot shows some considerable deviation from a normal distribution. This could be fixed by some transformations.
# In our previous price analysis, we saw that house prices observed positive skew
# We can achieve a more normal distribution if we convert prices to log(prices)
```

```python
df_last_25_years['log_prices'] = np.log(df_last_25_years['price'])
df_last_25_years['log_prices'].hist(figsize = (15,8))
```

```python
outcome = 'log_prices'
# note that we have changed our target variable into the logarithm of prices
x_cols =  ['sqft_living', 'waterfront', 'view', 'lat', 'long']
predictors = '+'.join(x_cols)
formula = outcome + '~' + predictors 
model2 = ols(formula = formula, data = df_last_25_years).fit()
display(model2.summary())

residuals = model2.resid
fig = sm.graphics.qqplot(residuals, dist=stats.norm, line='45', fit=True)
```

Our model seems to have improved considerably after transforming prices with a logarithmic function.

```python
df_last_25_years[['sqft_living', 'waterfront', 'view', 'lat', 'long']].hist(figsize = (15,8))
```

The distribution of sqft_living also shows positive skew, which can be transformed with a log function.

```python
df_last_25_years['log_sqft_living'] = np.log(df_last_25_years['sqft_living'])

log_sqft_living = df_last_25_years['log_sqft_living']

standardised_log_sqft_living = (log_sqft_living - np.mean(log_sqft_living))/np.sqrt(np.var(log_sqft_living))

display(standardised_log_sqft_living.hist(figsize = (10,6)))

```

```python
outcome = 'log_prices'
# again using the logarithm of prices
x_cols =  ['log_sqft_living', 'waterfront', 'view', 'lat', 'long']
predictors = '+'.join(x_cols)
formula = outcome + '~' + predictors 
model3 = ols(formula = formula, data = df_last_25_years).fit()
display(model3.summary())

residuals = model3.resid
fig = sm.graphics.qqplot(residuals, dist=stats.norm, line='45', fit=True)
```

Having now also used the logarithm of sqft_living, our new regression actually lost some predictive power in that the R squared value has dropped from 0.752 to 0.732. However we found that all other variables now have a stronger relationship in predicting house prices. Coefficients for waterfront, view and latitude all increased, but decreased for longitude.
